# Define Ingestion Configuration

```python
INGESTION_CONFIG = [
{
    "source":   "crm",
    "path":     "/Volumes/bikescatalog/bronze/raw_sources/engineering/source_crm/cust_info.csv",
    "table":    "crm_cust_info"
},
{
    "source":   "crm",
    "path":     "/Volumes/bikescatalog/bronze/raw_sources/engineering/source_crm/prd_info.csv",
    "table":    "crm_prd_info"
},
{
    "source":   "crm",
    "path":     "/Volumes/bikescatalog/bronze/raw_sources/engineering/source_crm/sales_details.csv",
    "table":    "crm_sales_details"
},
{
    "source":   "erp",
    "path":     "/Volumes/bikescatalog/bronze/raw_sources/engineering/source_erp/CUST_AZ12.csv",
    "table":    "erp_CUST_AZ12"
},
{
    "source":   "erp",
    "path":     "/Volumes/bikescatalog/bronze/raw_sources/engineering/source_erp/LOC_A101.csv",
    "table":    "erp_LOC_A101"
},
{
    "source":   "erp",
    "path":     "/Volumes/bikescatalog/bronze/raw_sources/engineering/source_erp/PX_CAT_G1V2.csv",
    "table":    "erp_PX_CAT_G1V2"
}

]
```

# Ingest files into bronze tables

```python
for item in INGESTION_CONFIG:
  print(f"Ingesting {item['source']} -> bikescatalog.bronze.{item['table']}")

  df = (
    spark.read.option("header", "true").option("inferSchema","true").csv(item["path"])
  )
  (
    df.write.mode("overwrite").format("delta").saveAsTable(f"bikescatalog.bronze.{item['table']}")
  )
```
